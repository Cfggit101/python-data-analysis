```python
df = pd.read_csv("tips.csv")
```

```python
#gonna try to grab last four digits of CC
#we can see CC is an interger, so i cant index last four digits of an interger
#but if i pass the int into a string, i can start to grab indexes:
str(1234567) [-4:]
#the slicing notation works on lists but also on STRINGS!
#the apply method is gonna allow me to create a custom function and apply it to a pandas series
#so lets functionalize this
```

```python
def last_four(num):
    return str(num)[-4:]
#gotta make sure the type of data i expect is the same that actually exists in the df
```

```python
last_four(567890)
#and so i wanna apply it to my CC column
```

```python
nums = df["CC Number"].apply(last_four)
nums
#just gonna pass the function, not gona call nor execute the function! cos pandas will do it, will grab the function and start applying it to every row of the column
```

```python
df["invented"] = df["day"]+nums
df
```

```python
#gonna grab the 'total bill' column and depending on the amount, gonna imput $, $$ or $$$ - like a cathegorical feature
```

```python
df["total_bill"].mean()
```

```python
def yelp(price):
    if price < 10:
        return("$")
    elif price >= 10 or price < 30:
        return("$$")
    else:
        return("$$$")
#remember its better to return values in my function, not print them directly so when i test it i can compare it with a test
#and remember to cover all-scenarios --> '=10' can also be!!!
```

```python
df["yelp"] = df["total_bill"].apply(yelp)
df
```
