```python
#HOW TO CHECK AND SELECT FOR MISSING VALUES:
```

```python
df = pd.read_csv('movie_scores.csv')
df
```

```python
df.isnull()
#returns true or false depending on whether it is a null vlaue or not --> True when it is, and False when not
```

```python
#to check for the opposite: 
df.notnull()
#False when not a value, true when value
```

```python
#And so with this, we can start to select columns and rows depending on whether its missing a value or not
```

```python
#Slect actors where pre-score its not missing: 
df['pre_movie_score'].notnull()
```

```python
#passing it to df:
df[df['pre_movie_score'].notnull()]
#will only get back the rows where i have pre-score value
```

```python
#so i can use these .isnull() and .notnull() as conditional filtering to select columns where certain features are present
```

```python
#pre-score missing and sex not missing:
df[(df['pre_movie_score'].isnull()) & (df['sex'].notnull())]
#remember! to combine conditions i use &!!!
#carefull, when i want the value to be there i use notnull and if i dont want it to be tere isnull, so that they return True (each one with its criteria), and return only the ones i want (that meet those requirements )
```

```python
#KEEP THE DATA - just read in the dataset an keep the missing values
```

```python
#DROP THE DATA
help(df.dropna)
#gives examples and the key paremeters:
    #axis --> whether you wanna drop columns or rows
    #how --> how you wanna drop them - 'any' is if any NA values present, drop that row or column; and 'all' is if all values are NA. drop that row or column
    #thres --> its gonna require that many non-null values
    #subset --> gonna allow you to only consider certain columns
```

```python
df.dropna()
#gonna drop any rows that have any missing values
```

```python
df['nan'] = [np.nan, np.nan, np.nan, np.nan, np.nan]
df
```

```python
df.dropna()
#if i had that column w nan, all rows would be dropped cos i have at least one nan value in each
```

```python
df.dropna(thresh=6)
#ganna keep the rows in which i have at least 6 not null values
```

```python
df.dropna(axis=1,thresh=4)
#gonna keep the columns with at least 4 not null values
```

```python
df.dropna(thresh=1)
#nice way of removing rows that have no values
```

```python
df.dropna(subset=['sex','pre_movie_score'])
#subset : column label or sequence of labels, optional
#Labels along other axis to consider, e.g. if you are dropping rows these would be a list of columns to include.
```

```python
#FILL IN MISSING DATA:
```

```python
help(df.fillna)
```

```python
#can use it to provide a single vlaue to fill in all missing values
df.fillna('NEW VALUE!')
#filling w a string its not a good idea cos its gonna transform numeric values such as the 63 into string when you probably wanna keep it as a numeric value
```

```python
#we can grab a column that we are interested in filling:
df['pre_movie_score'] = df['pre_movie_score'].fillna(0)
#reassign it to THE COLUMN to make it permanent
df
```

```python
#to fill in not by a prefixed value but by a function f ex the mean:
df['post_movie_score'] = df['post_movie_score'].fillna(df['post_movie_score'].mean())
df
```

```python
#if i got purely numeric column, i can directly say
df.fillna(df.mean())
#doesnt work now cos i have no more numeric columns to fill, but this would calculate the mean of the column and fill the NaN 
#and will leave non numerical columns as NaN
#dont know why it doesnt work on age one)
```
