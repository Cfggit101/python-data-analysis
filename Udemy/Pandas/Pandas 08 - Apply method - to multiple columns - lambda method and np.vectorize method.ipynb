```python
df = pd.read_csv("tips.csv")
```

```python
#LAMBDA EXPRESSIONS
#a lambda expression is an annonimous expression, meaning we dont actually apply a name to it in the same way we apply 
#a name to a function, we simply remove the 'def' and the name of the function, and replace it w 'lambda' and the input
#and delete 'return' and put it all in one line:
#def simple (num):
    #return num*2
#i turn it into:
lambda num: num*2
#lambda takes on a value and returns some operations on that
#not everything is convertible to a lambda expression!!! (f ex complex functions)
```

```python
df["total_bill"].apply(lambda num: num*2)
#we use lambda expression as a quick anonymous expression that you only intend to use one time
```

```python
def quality(total_bill,tip):
#its recomendable to use variables q same name as the columns - it'll help you keep track of things --> name them after the columns you intend to pass in
    if tip/total_bill > 0.25:
        return "Generous"
    else:
        return "Other"
```

```python
quality(16.99,1.01)
```

```python
#i gotta format it as a lambda expression:
df["quality"] = df[["total_bill", "tip"]].apply(lambda df: quality(df["total_bill"], df["tip"]),axis=1)
#1º grab the columns i'm gonna be using - pass a list of them
#2º call the apply method
#3º pass 'total_bill' and 'tip' into the quality function and im doing this calling 'lambda df'
#4º as a second argument to the apply mthod i write 'axis=1'
#--> my first parameter is the lambda expression an the second is 'axis=1'
df
```

```python
#--> you choose the columns you're gonna be using in your function, you call apply methond and lambda on the df being passed in, and then into your custom function (quality in this case), you pass the columns you use and lastly specify the axis equal to one 
#SUMMARY --> do a sub-selection of the columns you want. Call apply. Lambda DF. And then with your custom function name, pass in the columns you're using. And lastly, specify axis is equal to one.
```

```python
#TO MAKE THIS RUN FASTER --> NP.VECTORIZE
```

```python
df["Quality"] = np.vectorize(quality)(df["total_bill"],df["tip"])
#i just call vectorize, add the function i wanna vectorize and then call the two columns i wanna use in the function (or in which i wanna use the function)
```

```python
#whats does vectorize do? and why is it faster?:
#bc the purpose of np.vectorize is to transform functions which are not np aware --> its like i'm making the function aware thats gonna be used in a numpy array (and as so a pandas series)
#expl prof:
#If we take a look back at this original quality function, it is not aware that we're gonna be passing in and broadcasting this to what 
#is essentially a NumPy array within the Pandas series objects.
#So recall, every Pandas column in a data frame is a panda series, which in turn is holding a NumPy array. 
#And we know that if I broadcast specific NumPy aware methods like .mean or something on an NumPy array, it can perform this in a very efficient manner.
#But right now I have what is essentially what I'm gonna call a standard Python function. 
#It just accepts normal integers and floating point numbers. It's not actually thinking of this as being cast or broadcasted to a NumPy array.
#So if I actually transform this or vectorize this to be NumPy aware, then it actually will get computationally more efficient, 
#not because of np.vectorized, so to speak, but just because I'm really transforming this from a normal function to then be aware of a NumPy array.
#So np.vectorize won't actually improve performance if you're already dealing with something that's aware that it's taking in a NumPy array.
#But I have kind of this special case where this function is not aware that it's gonna be taking in a NumPy array. 
#In fact, it's not even designed to really think of that that way, it's just taking in single values. So by vectorizing it, 
#we actually are going to make this faster:
```
