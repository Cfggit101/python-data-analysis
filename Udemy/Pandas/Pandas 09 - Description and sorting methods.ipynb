```python
#SWITCHING OUT OR REPLACING VALUES - can be done through 'replace method' and 'map functionality'
```

```python
#replace method
#a single value:
df["sex"].replace("Female","F")
#takes in what you wanna replace, and what's the value you wanna replace it with
```

```python
#various colums: i enter them as a list - make sure the len of both list are the same (the ones you wanna replace and the ones for replacing), and that they are in the same order (order that you intend)
df["sex"].replace(["Female","Male"],["F","M"])
#recommendad to replace single values (f ex 'null' for '0')
```

```python
#map functionality
#you create a mapping as a dictionary
mymap = {"Female": "F", "Male":"M"}
df["sex"].map(mymap)
```

```python
#or: 
df["sex"].map({"Female": "F", "Male":"M"})
#itll map those values to the matching values 
#recommended for a lot of items - its more readable
```

```python
#BTW METHOD CALL
#will tell you which values in a particular column are in btw two values you choose
df["total_bill"].between(10,20)
#if i want it to be inclusive (the two numbers i chose included) --> 'inclusive=True' (dont know why it doesnt work)
```

```python
#HOW TO GRAB NLARGEST AND NSMALLEST
df.nlargest(2,"tip")
#Return the first `n` rows ordered by `columns` in descending order.
#this f ex returns me the top 10 tips
```

```python
#its like applying sort_value and then grabbing the fisrt two, or calling head of the top 2
df.sort_values("tip",ascending=False).iloc[0:2]
```

```python
df.nsmallest(2,"tip")
```

```python
#same as:
df.sort_values("tip",ascending=True).iloc[0:2]
```

```python
#FOR SAMPLING
#to grab a random sample of a df
#you can either sample back a certain number of rows or sample back a fraction of your dataframe
#certain number:
df.sample(5)
#each time i get different ones cos its a sample
```

```python
#fraction:
df.sample(frac=0.1)
#will fraction me a 10% of my dataframe -- useful so i dont have to know how long my rows in the dataframeare
#will grab them randomly too
```
