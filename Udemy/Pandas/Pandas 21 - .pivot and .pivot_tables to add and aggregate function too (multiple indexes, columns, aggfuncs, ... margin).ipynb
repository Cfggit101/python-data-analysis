.PIVOT_TABLE 
Pandas also comes with a pivot_table method that allows for an additional aggregation function to be called.
(In a lot of cases can be done easier w the .groupby)

```python
df = pd.read_csv('Sales_Funnel_CRM.csv')
df
```

```python
help(pd.pivot)
```

```python
#Gonna see how many licenses of each product type did Google purchase?
#1ยบ select the three or fez columns you wanna work with - company, licenses and product
licenses = df[['Company','Product','Licenses']]
```

```python
#2ยบ lets pivot
pd.pivot(data= licenses, columns= 'Product', index= 'Company', values= 'Licenses')
#got a lo to f NaN cos we havent sold every product to every company
```

```python
#now we gonna call an aggregation function on top of that
#how much have i sold to each company
#could do it w group by too
```

```python
df = pd.read_csv('Sales_Funnel_CRM.csv')
```

```python
df.groupby('Company').sum()['Sale Price']
```

```python
#index it kinda what you are actually gonna group by
pd.pivot_table(df,index='Company',  aggfunc = 'sum')
#pass the agregation function AS A STRING as another element in the pivot_table function
#if i dont pass the values element, itll perform the sum in all numeric values (acc number doesnt make much sense)
#can also see that it "groups", like "sums" the names of the persons i sold it to
```

```python
#to specify in wich ones you wanna sum - pass a list of columns:
pd.pivot_table(df,index='Company',  aggfunc = 'sum', values = ['Licenses','Sale Price'])
```

```python
df_google = df[df['Company'].isin([' Google'])]
df_google['Sale Price'].sum()
```

```python
#WITH TWO COLUMNS AS INDEX:
#to see acc manager with every contact he has and the sales he has w that contact:
df_accman = pd.pivot_table(df, index = ['Account Manager', 'Contact'], aggfunc= 'sum', values= ['Licenses','Sale Price'])
df_accman
#order is important! the first i pass in will be the outter index and the second the inner index
```

```python
#the agg method will be applied to the columns that you list, so i can call another column and so itll first separate by 'values', an dthen inside that, by the said column: - breaks them down
df_accmang = pd.pivot_table(df, index = ['Account Manager', 'Contact'], aggfunc= 'sum', columns= 'Product', values= ['Licenses','Sale Price'])
df_accmang
```

```python
#theres an automatic way of filling NaN values in pivot_tables -- add 'fill_value' and inidcate the value you want 
df_accmang = pd.pivot_table(df, index = ['Account Manager', 'Contact'], aggfunc= 'sum', fill_value= 0, columns= 'Product', values= ['Licenses','Sale Price'])
df_accmang
```

```python
df_accman['Sale Price'].sum()
```

```python
df['Sale Price'].sum()
```

```python
#TO CALL MULTIPLE AGGREGATION FUNCTIONS:
#call them as np.xxx and i can then pass a list
df_accmang = pd.pivot_table(df, index = ['Account Manager', 'Contact'], aggfunc= [np.sum, np.mean], fill_value= 0, columns= 'Product', values= ['Licenses','Sale Price'])
df_accmang
#will then bet all doubled w the mean too
```

```python
#to not mess things out, could put product instead of as a column, as an index too:
df_accmang = pd.pivot_table(df, index = ['Account Manager', 'Contact', 'Product'], aggfunc= [np.sum, np.mean], fill_value= 0, values= ['Licenses','Sale Price'])
df_accmang
```

```python
#PARAMETER 'margins
df_accmang = pd.pivot_table(df, index = ['Account Manager', 'Contact', 'Product'], aggfunc= [np.sum, np.mean], fill_value= 0, values= ['Licenses','Sale Price'], margins= True)
df_accmang
#True - gives you a grand total at the end of the margin (down)
#Could be usefull f ex to verify
```
