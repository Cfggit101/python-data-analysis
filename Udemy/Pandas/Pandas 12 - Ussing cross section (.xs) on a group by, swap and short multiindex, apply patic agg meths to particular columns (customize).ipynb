```python
#USSING CROSS SECTION - grabbing smth that has this multiplevel index --> using .xs()
#Takes in a key, index (to look at rows '0' cos we have our multi index in rows), and level (what level of index youre looking at (inner or outter in this case))
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('mpg.csv')
df
```

```python
year_cyl = df.groupby(['model_year', 'cylinders']).count()
year_cyl
```

```python
#to grab all from 1970
year_cyl.xs(key=70,level='model_year')
#'level' is like at what level im looking at 
#keep in mind it'll only take one key value!
```

```python
#to take based on inner index:
year_cyl.xs(key=4,level='cylinders')
#i take 4 cyls for every single year

#keep in mind the level i ask for wont appear in the answer cos they are all from the same year or cylinders f ex
#so best to put in variables name what it is cos it wont tell in the answer (output)
```

```python
#if you want from two keys:
#as you cannot put more that one key --> better to filter those values im interested in BEFORE the group by 
dff = df[df['cylinders'].isin([6,8])].groupby(['model_year','cylinders']).count()
dff
# so 1Âº i choose only the 6 and 8s in the df (filter out df first)
#2Âº i group it as i want
```

```python
dff
```

```python
#TO SWAP THE MULTIINDEX
dff.swaplevel()
```

```python
#TO SORT THE MULTIINDEX
dff.sort_index(level='model_year',ascending=False)
#same as before but i specify the level i wanna sort
```

```python
#keep in mind that if i sort the inner one differently maybe ill "break" the outter and make it more confussing
dff.sort_index(level='cylinders', ascending=False)
#so it makes more sense to sort by the outter level
```

```python
df['cylinders'].transform(pd.to_numeric, errors = 'coerce')
df['displacement'].transform(pd.to_numeric, errors = 'coerce')
df['horsepower'].transform(pd.to_numeric, errors = 'coerce')
df['weight'].transform(pd.to_numeric, errors = 'coerce')
df['acceleration'].transform(pd.to_numeric, errors = 'coerce')
df['mpg'].transform(pd.to_numeric, errors = 'coerce')
```

```python
#CUSTOMIZE AGGREGATE FUNCTIONS PER CATHEGORY - to not make same aggregate function in every single column
# i specify the aggregate function for every column
df.agg(['std','mean'])['mpg']
#takes in a list of string calls
#SHOULD WORK
```

```python
#APPLY PARTICULAR AGGREGATION METHODS TO PARTICULAR COLUMNS
#or to apply certain aggregate functions to specific columns - i use the form of a dictionary
#the key is the column i wanna apply it in and the value is the aggregate function i wanna apply (can pass more than one)
dfff = df.agg({'mpg':['max','mean'], 'weight':['mean','min','std']})
dfff
#keep in mid itll fill in w NaN the values you didnt call
```

```python
dfff['weight'] = dfff['weight'].fillna(dfff['weight'].max())
dfff
```
