```python
#GROUP BY OPERATIONS
# A groupby operation call allows us to examine data on a per category basis.
# f ex average per group or per category
#it ll aggregate a continuous value per category - w non continuous columns and have distinct categories (not continuous numeric ones)
#they dont always need to be strings - can also be numericals
#when we've chosen the column we want to group by, we call the group by function
#THIS GROUP BY FUNCTION HAS TO BE COMBINED W AN AGGREGATE FUNCTION CALL!!!
#itll separate the columns by the categories, and then aggregates them depending on the function you entered (average, medium, sum, count,...)

#So note that in pandas, calling just groupby by itself creates what's known as a lazy groupby object because it's waiting still to be evaluated by an aggregate method call.
# --> always needs a aggregate function
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('mpg.csv')
df
```

```python
#how performance has changed throughout the years -- see mpg through years
#takes knowledge of the df to know which column is categorical and which continuous
#here i choose mpg as column i wanna group by ('categorical column')
#i can use unique function to the unique values of the df:
df['model_year'].unique()
```

```python
#or value counts to see how many rows i have of each unique value
df['model_year'].value_counts()
#its kind of a very fast group by
```

```python
df.groupby('model_year').count()['mpg']
#if i leave it as only groupby, it wont do anaything cos itll be waiting for the aggregate function to tell him what to do
#MOST COMMON ONES: 
#mean(): Compute mean of groups
#sum(): Compute sum of group values
#size(): Compute group sizes
#count(): Compute count of group
#std(): Standard deviation of groups
#var(): Compute variance of groups
#sem(): Standard error of the mean of groups
#describe(): Generates descriptive statistics
#first(): Compute first of group values
#last(): Compute last of group values
#nth() : Take nth value, or a subset if n is a list
#min(): Compute min of group values
#max(): Compute max of group values

#So what it's going to do is it's going to grab the column you want, it's going to groupby that particular category, set that as the index, and then it takes in the aggregate of what it can do.
#to call just one i put it btw[]
```

```python
#TO GROUP BY MULTIPLE COLUMNS - MULTILEVEL OR HIERARCHY INDEX:
#i pass it a list of the columns - itll first group by the fisrt one and btw the ones that are the ame, itll use the next and the next
df.groupby(['model_year', 'cylinders']).count()
# so now modelyear and cylinders are not columns names, they are indexes - can see by putting at the end '.columns' - wont say those two
#this way i can compare btw cylinders  btw years
```

```python
df.groupby(['model_year', 'cylinders']).describe().transpose()
#shows aggregations possible for a single column - to grab simple statistical analysis of df
```

```python
df.groupby(['model_year', 'cylinders']).count().index
#it returns a multiindex as touples
```

```python
#HOW TO GRAB THINGS IN A MULTI-INDEX
year_cyl = df.groupby(['model_year','cylinders']).count()
year_cyl
```

```python
#can call indexes names:
year_cyl.index.names
```

```python
year_cyl.index.levels
#its a list of lists (MultiD)
#first one the years and second possible cylinders (keep in mind not every cylinder is present for every value)
```

```python
#TO GRAB BY THE OUTSIDE INDEX:
#can grab by that outside index using .loc
year_cyl.loc[70]
#will work on the most outside index
#drops the index name column
```

```python
year_cyl.loc[[70,82]]
#if i ask for two (passing it a list) itll put the indexes names in order to distinguis
```

```python
#TO GRAB A PARTICULAR ROW:
year_cyl.loc[(70,4)]
#gotta pass it the touple
```
