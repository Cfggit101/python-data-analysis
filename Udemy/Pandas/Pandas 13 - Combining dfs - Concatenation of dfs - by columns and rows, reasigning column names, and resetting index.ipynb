```python
#CONCATENATION - if both soures are already in the same format
#its like "pasting" two df together by columns - requirement is that they have rows in common
#can also be made on rows 
#will automatically fill w NaN where necessary
```

```python
import pandas as pd
import numpy as np
```

```python
d1 = {'A' : ['A0','A1','A2','A3'], 'B': ['B0','B1','B2','B3']}
```

```python
d2 = {'C' : ['C0','C1','C2','C3'], 'D': ['D0','D1','D2','D3']}
```

```python
one = pd.DataFrame(d1)
```

```python
two = pd.DataFrame(d2)
```

```python
one
```

```python
two
```

```python
res = pd.concat([one,two],axis=1)
#along columns - pass in a list the concatenation objets
#to concatenate along columns i specify 'axis=1'!!!
res
#if i pass 'two,one' they'd be switched (C,D,A,B)
```

```python
res = pd.concat([one,two],axis=0)
#fills w NaN
res
```

```python
#but if i wanna remane C and D as A and B so they are in the same columns (A and B):
#i remane the columns for them to match up (before the concatenation):
#i say columns for two are the same as columns for one:
two.columns = one.columns
#will remane the columns maintaining same inside values
two
```

```python
#and now i concatenate them:
res = pd.concat([one,two],axis=0)
res
```

```python
#to reset the index:
res.index = range(len(res))
res
```
