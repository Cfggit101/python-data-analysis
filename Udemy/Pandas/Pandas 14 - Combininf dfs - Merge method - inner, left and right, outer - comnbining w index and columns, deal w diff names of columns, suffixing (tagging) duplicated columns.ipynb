```python
import numpy as np
import pandas as pd
```

```python
#COMBINING DFS
#INNER MERGE METHOD
#when they are not in the same order or format 
#simmilar to 'join' command in SQL

# its pd.merge() - takes in a key argument 'how' - three ways of merging w how:
# 1) Inner
# 2) Outter
# 3) Left or Right

#The main idea is to decide how to deal w the info only present in one of the joined tables - you wanna keep the ones not commons (w NaNs) or you wanna drop them altogether

#Merging process:
# 1ยบ - decide on what columnt to merge together
    #the primary column should be a primary identifier - meaning unique per row
    # the on column should be present in both tables being merged - and represents the same thing in both tables
# 2ยบ - decide how to merge the tables on the elected column
    #w 'how' = 'inner' --> the result will be the set of records that match in both tables - ill only be looking at the ones that are present in both tables - like a venn diagram (only the area common to both circles)
    # would look like --> pd.merge(df1, df2, how = 'inner', on = 'on column') -- will return the common values of the 'on column' w the values from both dfs associated to them
```

```python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
registrations
```

```python
logins
#could reset the column to be the index if wanted
```

```python
help(pd.merge)
```

```python
pd.merge(registrations, logins, how = 'inner', on ='name')
#as you are taking values present in both tables, the order doesnt really matter (just in inner)
```

```python
#LEFT AND RIGHT MERGE
#order in which i pass in the df here DOES matter!!! - first i pass in is the left one and second the right one
#if i put 'how = 'left'' - itll return all values in the left one and only the common ones of the right one -- ill have NaN in the values present in left but not in right one
#in the 'right' situation will happen the opposite - will have all form right and only the common ones of left
```

```python
pd.merge(left = registrations, right = logins, how = 'left', on = 'name')
```

```python
df = pd.merge(left = registrations, right = logins, how = 'right', on = 'name')
df
#pandas is actually smart enough to know in this case which should be the 'on column', so i could delete that part, but better to clarify it so i can know my intentions when i wrote it
```

```python
df.index = df['name']
df
```

```python
#OUTTER MERGE:
#allows us to include everything present in both tables
pd.merge(registrations, logins, how='outer', on='name')
#the order of the dfs not important either
```

```python
#JOIN ON INDEX INSTEAD OF COLUMNS
registrations
#now i wanna merge on the index of registrations and the column 'name' in logins
#the merge function takes also a:
    #'left_index' (index of left table)  -- boolean
    #'right_index' (index of right table)   -- boolean; 
#parameters as well as a 
    #'left_on' (a column in the left table) -- enter name of the specific column
    #'right_on' (a column in the right table) -- enter name of the specific column
#we can combine this and use it:
```

```python
pd.merge(registrations, logins, how = 'inner', left_index = True , right_on = 'name' )
#in the column one i have to enter the name of the column
```

```python
#HOW TO DEAL W DIFFERENT NAMES OF THE 'ON COLUMN'
#when i wwanna merge two columns that mean the same but are called different
registrations = registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
registrations
```
